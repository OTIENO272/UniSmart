```python
import sys
from PyQt5 import QtCore, QtGui, QtWidgets
import mysql.connector
from PyQt5.QtWidgets import QMessageBox

class Ui_MainWindow(object):
    def __init__(self):
        self.current_index = -1 

    def setupUi(self, MainWindow):
        MainWindow.setObjectName("MainWindow")
        MainWindow.resize(1500, 918)
        MainWindow.setStyleSheet("""
            QMainWindow {
                background-color: #f4f7f6;
            }
            QLabel {
                color: #2c3e50;
                font-size: 14px;
                font-weight: bold;
            }
            QLineEdit {
                background-color: white;
                border: 2px solid #bdc3c7;
                border-radius: 6px;
                padding: 5px;
                color: #2c3e50;
            }
            QLineEdit:focus {
                border: 2px solid #3498db;
            }
            QPushButton {
                background-color: #2c3e50;
                color: white;
                border-radius: 6px;
                font-weight: bold;
                padding: 5px;
            }
            QPushButton:hover {
                background-color: #34495e;
            }
            QGroupBox {
                border: 2px solid #bdc3c7;
                border-radius: 8px;
                margin-top: 10px;
                font-weight: bold;
                color: #2c3e50;
            }
            QTableWidget {
                background-color: white;
                gridline-color: #ecf0f1;
                border: 1px solid #bdc3c7;
                selection-background-color: #3498db;
                color: #2c3e50;
            }
            QHeaderView::section {
                background-color: #2c3e50;
                color: white;
                padding: 4px;
                border: 1px solid #7f8c8d;
            }
        """)
        self.centralwidget = QtWidgets.QWidget(MainWindow)
        MainWindow.setWindowTitle("UniSmart")

        # --- LABELS ---
        # Adjusted positions and text to match "Second Name" and "Surname"
        self.lbl_reg = QtWidgets.QLabel("Registration Number:", self.centralwidget)
        self.lbl_reg.setGeometry(QtCore.QRect(250, 40, 200, 41))
        
        self.lbl_secondName = QtWidgets.QLabel("Second Name:", self.centralwidget)
        self.lbl_secondName.setGeometry(QtCore.QRect(250, 110, 200, 41))
        
        self.lbl_surname = QtWidgets.QLabel("Surname:", self.centralwidget)
        self.lbl_surname.setGeometry(QtCore.QRect(250, 190, 200, 41))
        
        self.lbl_course = QtWidgets.QLabel("Course:", self.centralwidget)
        self.lbl_course.setGeometry(QtCore.QRect(250, 280, 200, 41))
        
        self.lbl_county = QtWidgets.QLabel("County:", self.centralwidget)
        self.lbl_county.setGeometry(QtCore.QRect(250, 400, 200, 41))
        
        self.lbl_gender = QtWidgets.QLabel("Gender:", self.centralwidget)
        self.lbl_gender.setGeometry(QtCore.QRect(250, 510, 200, 41))
        
        self.lbl_dept = QtWidgets.QLabel("Department:", self.centralwidget)
        self.lbl_dept.setGeometry(QtCore.QRect(250, 630, 200, 41))
        
        self.lbl_school = QtWidgets.QLabel("School:", self.centralwidget)
        self.lbl_school.setGeometry(QtCore.QRect(250, 720, 200, 41))

        # --- INPUTS ---
        input_style = "background-color: white; color: black; border-radius: 4px;"
        
        self.regLineEdi = QtWidgets.QLineEdit(self.centralwidget)
        self.regLineEdi.setGeometry(QtCore.QRect(480, 39, 221, 41))
        self.regLineEdi.setStyleSheet(input_style)

        self.secondNameLineEdit_2 = QtWidgets.QLineEdit(self.centralwidget)
        self.secondNameLineEdit_2.setGeometry(QtCore.QRect(480, 109, 221, 41))
        self.secondNameLineEdit_2.setStyleSheet(input_style)

        self.surnameLineEdit = QtWidgets.QLineEdit(self.centralwidget)
        self.surnameLineEdit.setGeometry(QtCore.QRect(480, 190, 221, 41))
        self.surnameLineEdit.setStyleSheet(input_style)

        self.courseLineEdit = QtWidgets.QLineEdit(self.centralwidget)
        self.courseLineEdit.setGeometry(QtCore.QRect(480, 280, 221, 41))
        self.courseLineEdit.setStyleSheet(input_style)

        self.countyLineEdit = QtWidgets.QLineEdit(self.centralwidget)
        self.countyLineEdit.setGeometry(QtCore.QRect(480, 400, 221, 41))
        self.countyLineEdit.setStyleSheet(input_style)

        self.deparmenLineEdit = QtWidgets.QLineEdit(self.centralwidget)
        self.deparmenLineEdit.setGeometry(QtCore.QRect(480, 629, 231, 41))
        self.deparmenLineEdit.setStyleSheet(input_style)

        self.schoolLineEdit = QtWidgets.QLineEdit(self.centralwidget)
        self.schoolLineEdit.setGeometry(QtCore.QRect(480, 720, 231, 41))
        self.schoolLineEdit.setStyleSheet(input_style)

        # Gender
        self.genderGroupBox = QtWidgets.QGroupBox("GENDER", self.centralwidget)
        self.genderGroupBox.setGeometry(QtCore.QRect(480, 480, 231, 101))
        self.radioButton = QtWidgets.QRadioButton("MALE", self.genderGroupBox)
        self.radioButton.setGeometry(QtCore.QRect(20, 30, 111, 28))
        self.radioButton_2 = QtWidgets.QRadioButton("FEMALE", self.genderGroupBox)
        self.radioButton_2.setGeometry(QtCore.QRect(20, 60, 111, 28))

        # --- BUTTONS ---
        btn_style = "font-weight: bold;"
        self.addButton = QtWidgets.QPushButton("Add Record", self.centralwidget)
        self.addButton.setGeometry(QtCore.QRect(20, 788, 100, 41))
        self.addButton.setStyleSheet(btn_style)
        
        self.updateButton = QtWidgets.QPushButton("Update Record", self.centralwidget)
        self.updateButton.setGeometry(QtCore.QRect(140, 788, 120, 41))
        
        self.DeleteButton = QtWidgets.QPushButton("Delete Record", self.centralwidget)
        self.DeleteButton.setGeometry(QtCore.QRect(280, 788, 120, 41))
        
        self.SearchButton = QtWidgets.QPushButton("Search Record", self.centralwidget)
        self.SearchButton.setGeometry(QtCore.QRect(480, 790, 120, 41))
        
        self.ResetBuon = QtWidgets.QPushButton("Reset Form", self.centralwidget)
        self.ResetBuon.setGeometry(QtCore.QRect(650, 788, 120, 41))
        
        self.Next = QtWidgets.QPushButton("Next >", self.centralwidget)
        self.Next.setGeometry(QtCore.QRect(810, 788, 100, 41))
        
        self.Previous = QtWidgets.QPushButton("< Previous", self.centralwidget)
        self.Previous.setGeometry(QtCore.QRect(930, 788, 100, 41))

        # --- TABLE ---
        self.tableWidget = QtWidgets.QTableWidget(self.centralwidget)
        self.tableWidget.setGeometry(QtCore.QRect(750, 40, 720, 720))
        self.tableWidget.setColumnCount(8)
        self.tableWidget.setHorizontalHeaderLabels(["Reg No", "Second Name", "Surname", "Course", "County", "Gender", "Dept", "School"])
        self.tableWidget.setStyleSheet("background-color: white; color: black;")

        MainWindow.setCentralWidget(self.centralwidget)

        # --- CONNECTIONS ---
        self.addButton.clicked.connect(self.add_record)
        self.updateButton.clicked.connect(self.update_record)
        self.DeleteButton.clicked.connect(self.delete_record)
        self.SearchButton.clicked.connect(self.search_record)
        self.ResetBuon.clicked.connect(self.reset_form)
        self.Next.clicked.connect(self.move_next)
        self.Previous.clicked.connect(self.move_previous)
        
        self.load_table()

    # (Database methods: get_db_connection, load_table, add_record, update_record, delete_record, search_record, display_record, move_next, move_previous, select_row_from_index, reset_form - keep these from the previous code block)

    def get_db_connection(self):
        return mysql.connector.connect(host="localhost", user="root", password="@Vich272", database="University")

    def load_table(self):
        try:
            conn = self.get_db_connection()
            cursor = conn.cursor()
            cursor.execute("SELECT * FROM Students")
            records = cursor.fetchall()
            self.tableWidget.setRowCount(0)
            for row_num, row_data in enumerate(records):
                self.tableWidget.insertRow(row_num)
                for col_num, data in enumerate(row_data):
                    self.tableWidget.setItem(row_num, col_num, QtWidgets.QTableWidgetItem(str(data)))
            conn.close()
        except Exception as e: print(f"Load Error: {e}")

    def add_record(self):
        gender = "MALE" if self.radioButton.isChecked() else "FEMALE"
        query = "INSERT INTO Students VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
        data = (self.regLineEdi.text(), self.secondNameLineEdit_2.text(), self.surnameLineEdit.text(),
                self.courseLineEdit.text(), self.countyLineEdit.text(), gender, 
                self.deparmenLineEdit.text(), self.schoolLineEdit.text())
        try:
            conn = self.get_db_connection()
            cursor = conn.cursor()
            cursor.execute(query, data)
            conn.commit()
            conn.close()
            self.load_table()
            QMessageBox.information(None, "Success", "Record Added!")
        except Exception as e: QMessageBox.critical(None, "Error", str(e))

    def update_record(self):
        gender = "MALE" if self.radioButton.isChecked() else "FEMALE"
        query = "UPDATE Students SET SecondName=%s, Surname=%s, Course=%s, County=%s, Gender=%s, Department=%s, School=%s WHERE RegistrationNumber=%s"
        data = (self.secondNameLineEdit_2.text(), self.surnameLineEdit.text(), self.courseLineEdit.text(), 
                self.countyLineEdit.text(), gender, self.deparmenLineEdit.text(), 
                self.schoolLineEdit.text(), self.regLineEdi.text())
        try:
            conn = self.get_db_connection()
            cursor = conn.cursor()
            cursor.execute(query, data)
            conn.commit()
            conn.close()
            self.load_table()
            QMessageBox.information(None, "Updated", "Record modified successfully!")
        except Exception as e: QMessageBox.critical(None, "Error", str(e))

    def delete_record(self):
        reg = self.regLineEdi.text()
        if not reg: return
        try:
            conn = self.get_db_connection()
            cursor = conn.cursor()
            cursor.execute("DELETE FROM Students WHERE RegistrationNumber=%s", (reg,))
            conn.commit()
            conn.close()
            self.load_table()
            self.reset_form()
            QMessageBox.information(None, "Deleted", "Record removed.")
        except Exception as e: QMessageBox.critical(None, "Error", str(e))

    def search_record(self):
        reg = self.regLineEdi.text()
        try:
            conn = self.get_db_connection()
            cursor = conn.cursor()
            cursor.execute("SELECT * FROM Students WHERE RegistrationNumber=%s", (reg,))
            row = cursor.fetchone()
            if row:
                self.display_record(row)
            else:
                QMessageBox.warning(None, "Not Found", "No student with that Reg Number.")
            conn.close()
        except Exception as e: print(e)

    def display_record(self, row):
        self.regLineEdi.setText(str(row[0]))
        self.secondNameLineEdit_2.setText(str(row[1]))
        self.surnameLineEdit.setText(str(row[2]))
        self.courseLineEdit.setText(str(row[3]))
        self.countyLineEdit.setText(str(row[4]))
        if str(row[5]).upper() == "MALE": self.radioButton.setChecked(True)
        else: self.radioButton_2.setChecked(True)
        self.deparmenLineEdit.setText(str(row[6]))
        self.schoolLineEdit.setText(str(row[7]))

    def move_next(self):
        count = self.tableWidget.rowCount()
        if count > 0:
            self.current_index = (self.current_index + 1) % count
            self.select_row_from_index()

    def move_previous(self):
        count = self.tableWidget.rowCount()
        if count > 0:
            self.current_index = (self.current_index - 1) % count
            self.select_row_from_index()

    def select_row_from_index(self):
        row_data = []
        for i in range(8):
            item = self.tableWidget.item(self.current_index, i)
            row_data.append(item.text() if item else "")
        self.display_record(row_data)
        self.tableWidget.selectRow(self.current_index)

    def reset_form(self):
        for widget in self.centralwidget.findChildren(QtWidgets.QLineEdit):
            widget.clear()
        self.current_index = -1

if __name__ == "__main__":
    app = QtWidgets.QApplication(sys.argv)
    MainWindow = QtWidgets.QMainWindow()
    ui = Ui_MainWindow()
    ui.setupUi(MainWindow)
    MainWindow.show()
    sys.exit(app.exec_())
```
